```python
import pandas as pd
#importing more of python libraries like numpy, matplot, seaborn and scipy with pandas and sklearn.
import numpy as np
```

```python
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt

import seaborn as sns

from scipy import stats

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
#Print the first 5 rows to observe your data
iris_df.head()
```

1. Compute mean, median, and mode for each feature.

```python
print("\nMean of each feature:")
print(iris_df.mean())

print("\nMedian of each feature:")
print(iris_df.median())

print("\nMode of each feature:")
print(iris_df.mode().iloc[0])
```

2. Calculate the range, variance, and standard deviation for each feature.

```python
print("\nRange of each feature:")
print(iris_df.max() - iris_df.min())

print("\nVariance of each feature:")
print(iris_df.var())

print("\nStandard deviation of each feature:")
print(iris_df.std())
```

3. Compute the 25th, 50th, and 75th percentiles (quartiles) for each feature.

```python
print("\n25th percentile of each feature:")
print(iris_df.quantile(0.25))

print("\n50th percentile of each feature (Median):")
print(iris_df.quantile(0.5))

print("\n75th percentile of each feature:")
print(iris_df.quantile(0.75))
```

4. Create a histogram for each of the features to visualize their distributions.

```python
iris_df.hist(bins=20, figsize=(10, 10))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()
```

5. Plot scatter plots between pairs of features to explore relationships.

```python
sns.pairplot(iris_df)
plt.suptitle("Scatter Plots between Pairs of Features")
plt.show()
```
