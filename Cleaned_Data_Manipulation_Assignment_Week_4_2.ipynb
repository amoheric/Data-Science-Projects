Download the dataset and load it as a pandas dataframe and perform the following:

```python
import pandas as pd

# Attempting to load the dataset from the file path
file_path = '/content/Fast-Food Chains in USA.csv'
df = pd.read_csv(file_path)

# Displaying the structure of the dataset
print(df.info())
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python

# Cleaning the data by removing rows with null/NAN values
df = df.dropna()

# Displaying the structure of the cleaned dataset
print(df.info())
print("\nFirst few rows of the cleaned dataset:")
print(df.head())
```

```python


# Display the columns to identify the correct column name
print("Columns in the dataset:")
print(df.columns)
```

1. Using pandas to insert a new record into the dataset with the following details:

Fast-Food Chain: "Subway",

Systemwide Sales: $1000 million,

Average Sales per Unit: $400 thousand,

Franchised Stores: 2000,

Company Stores: 500,

2021 Total Units: 2500,

Total Change in Units from 2020: 50

```python
new_record = {
    'Fast-Food Chain': 'Subway',
    'Systemwide Sales': 1000,
    'Average Sales per Unit': 400,
    'Franchised Stores': 2000,
    'Company Stores': 500,
    '2021 Total Units': 2500,
    'Total Change in Units from 2020': 50
}
```

```python

# Converting the new_record to a DataFrame and concatenate with the existing DataFrame
new_record_df = pd.DataFrame([new_record])
df = pd.concat([df, new_record_df], ignore_index=True)

# Displaying the updated DataFrame
print(df)
```

2. Update the systemwide sales for the fast food chain with the name "McDonald's" to $40000 million.

```python
df.loc[df['Fast-Food Chain'] == "McDonald's", 'Systemwide Sales'] = 40000
```

```python

# Checking if the 'Category' column exists, if not, identify the correct column name
print("Columns in the dataset after inserting new record:")
print(df.columns)
```

3. Using pandas to calculate the average systemwide sales for each category of fast food chains, grouped by category.

```python
avg_sales_by_category = df.groupby('Fast-Food Chain')['Systemwide Sales'].mean()

print("Average systemwide sales by category:")
print(avg_sales_by_category)
```

4.Using pandas retrieve the top 10 fast food chains with the highest average sales per unit, ordered by average sales per unit in descending order.

```python
top_10_chains_by_avg_sales = df.nlargest(10, 'Average Sales per Unit')[['Fast-Food Chain', 'Average Sales per Unit']]

print("\nTop 10 fast food chains by average sales per unit:")
print(top_10_chains_by_avg_sales)
```

5. Using pandas retrieve the names of all fast food chains that have more franchised stores than company stores.

```python
chains_more_franchised_than_company = df[df['Franchised Stores'] > df['Company Stores']]['Fast-Food Chain']

print("\nFast food chains with more franchised stores than company stores:")
print(chains_more_franchised_than_company)
```

6. Using pandas delete all records from the dataset where the total change in units from 2020 is negative.

```python
df = df[df['Total Change in Units from 2020'] >= 0]

print("\nDataset after deleting records with negative change in units from 2020:")
print(df)
```

```python

# Displaying all my concluded data manipulation results
print("Average systemwide sales by category:")
print(avg_sales_by_category)

print("\nTop 10 fast food chains by average sales per unit:")
print(top_10_chains_by_avg_sales)

print("\nFast food chains with more franchised stores than company stores:")
print(chains_more_franchised_than_company)

print("\nDataset after deleting records with negative change in units from 2020:")
print(df)
```
