1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
Were the passengers in higher classes more likely to survive?

```python
#Code here

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Titanic dataset
file_path = '/content/Titanic.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the column names to verify
print("Columns in the dataset:")
print(df.columns)

# Ensuring my columns are correctly named and exist in my dataset
# 1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
# Were the passengers in higher classes more likely to survive?
sns.catplot(x="class", hue="survived", kind="count", data=df)
plt.title('Survival Rate by Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Count')
plt.show()

```

Explanation here:
<br>

This plot shows that passengers in higher classes had higher survival rates compared to lower classes.


<br>

<br>

2. Create a histogram to visualize the distribution of passenger ages. What is the age distribution like on the Titanic?

```python
#Code here

sns.histplot(df['age'].dropna(), kde=False, bins=30)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Explanation here:

The age distribution on the Titanic shows a higher frequency of passengers in the 20-40 age range.



<br>

<br>

3. Generate a bar chart to compare the survival rates of male and female passengers. Did gender play a role in survival?

```python
#Code here

sns.catplot(x="sex", hue="survived", kind="count", data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Explanation here:

The bar chart indicates that female passengers had a higher survival rate compared to male passengers.



<br>

<br>


4. Visualize the distribution of ticket fares for each passenger class using box plots. How did ticket fares vary across different classes?

```python
# Code here
sns.boxplot(x="class", y="fare", data=df)
plt.title('Distribution of Ticket Fares by Passenger Class')
plt.xlabel('Passenger Ticket Class')
plt.ylabel('Ticket Fare')
plt.show()
```

Explanation here:

The box plot shows that ticket fares were highest for 1st class passengers, with a wider range of fares compared to 2nd and 3rd classes.


<br>

<br>


5. Create a grouped bar chart to visualize the survival rates of passengers across different age groups and genders. Were certain age groups or genders more likely to survive?

```python
# Code here
df['age_group'] = pd.cut(df['age'], bins=[0, 12, 18, 35, 60, 100], labels=['Child', 'Teenager', 'Adult', 'Middle-Aged', 'Senior'])
sns.catplot(x="age_group", hue="survived", col="sex", kind="count", data=df)
plt.show()
```

Explanation here:

This chart shows that certain age groups and genders had different survival rates, with children and females generally having higher survival rates.
